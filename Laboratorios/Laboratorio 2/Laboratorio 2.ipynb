### 1. ¿Cómo podemos detectar la fatiga muscular de manera objetiva usando señales EMG en lugar de evaluaciones subjetivas?
Las evaluaciones actuales dependen de autoinformes de dolor o cansancio, que son poco confiables. Estos métodos son subjetivos y pueden variar según la persona o el contexto cultural, por lo que no siempre muestran el estado real del músculo [1].  

La electromiografía de superficie (sEMG) ofrece una perspectiva más objetiva, ya que registra la actividad eléctrica de los músculos cuando se contraen. Estos registros permiten ver directamente cambios fisiológicos que ocurren durante la fatiga, como la disminución en la **frecuencia media (MNF)** y la **frecuencia mediana (MDF)**, el aumento del valor **RMS (root mean square)** debido al reclutamiento de más fibras musculares y las variaciones en la velocidad de conducción junto con la sincronización de las unidades motoras [2,3].  

Actualmente hay estudios que han validado el uso de sEMG para la detección de fatiga muscular. Se ha propesto un índice de fatiga basado en la diferencia entre frecuencias altas y bajas de la señal EMG, con una precisión del 94.66% en el músculo deltoides medio [4]. También, se aplicaron técnicas de aprendizaje automático en señales EMG de contracciones isométricas, logrando una precisión del 99.8% para clasificar estados de fatiga [5]. En un contexto educativo, el uso del kit **BITalino** en Perú permitió la clasificación de señales EMG en estados de fatiga y no-fatiga con una precisión del 90%, mostrando que este tipo de tecnologías también son viables para estas aplicaciones [1].    

---

### 2. ¿Qué características de la señal EMG son más relevantes para identificar el inicio de la fatiga?
Estudios previos han mostrado que la fatiga se refleja en 3 aspectos. Se observa un aumento en la amplitud de la señal, lo cual se mide a través de indicadores como RMS (root mean square), MAV (mean absolute value) o IEMG (integrated EMG). La amplitud aumenta a medida que el músculo se fatiga porque se reclutan más fibras musculares y disminuye la velocidad de conducción. Debido a esto, Qassim et al. propusieron un índice de fatiga basado en IEMG que logró detectar la progresión de la fatiga en el músculo deltoides con una precisión del 94.66% [4].

También aparece una disminución en las frecuencias medias del espectro de la señal que corresponden a MNF (mean frequency) y MDF (median frequency). Estos parámetros disminuyen casi linealmente durante contracciones prolongadas, lo que indica que la señal se desplaza hacia frecuencias más bajas conforme avanza la fatiga [6]. Se han propuesto razones como MNF/ARV (average rectified value) que permiten cuantificar de forma más sensible la fatiga muscular [7].

El tercer aspecto corresponde a cambios más complejos en el dominio del tiempo y frecuencia de la señal. Al ser más complejos no son siempre detectables con métricas simples. Wavelet Transform, ICEEMDAN o la entropía muestral (SampEn) permiten captar variaciones no lineales y transitorias. Sun et al. aplicaron ICEEMDAN combinado con clasificación SVM y lograron un 99.8% de precisión en la detección de fatiga muscular en contracciones isométricas [5].

---

### 3. ¿Es posible diferenciar entre contracciones normales y estados de fatiga muscular mediante un modelo de clasificación?

Sí se puede diferenciar entre reposo, contracción normal y fatiga muscular utilizando modelos de clasificación entrenados con características extraídas de la señal EMG. Estos modelos permiten desarrollar sistemas automáticos capaces de detectar de manera temprana la fatiga usando diferentes características de la señal EMG. En el dominio del tiempo: RMS, MAV e IEMG; en el dominio de la frecuencia: MNF y MDF y en el dominio de tiempo-frecuencia: transformadas wavelet o entropía muestral [2], [6,7].

El algoritmo Support Vector Machines (SVM) se ha combinado con métodos de descomposición de la señal como ICEEMDAN para identificar patrones de fatiga, alcanzando una precisión del 99.8% en contracciones isométricas de miembros inferiores [5]. De manera similar, se han probado modelos más simples como K-Nearest Neighbors (KNN), que ofrecen buenos resultados en condiciones con bajo nivel de ruido, y redes más complejas como las Convolutional Neural Networks (CNN), capaces de reconocer patrones más complejos en la señal [2]. El algoritmo Double-Step Binary Classifier analiza primero si existe contracción sostenida y luego clasifica entre componentes de alta y baja frecuencia para estimar la fatiga, logrando una precisión del 94.66% en el músculo deltoides medio [4].

---
### 4. ¿Qué tan factible es implementar un sistema de monitoreo en tiempo real basado en EMG en contextos laborales?
La implementación de sistemas sEMG en tiempo real en contextos laborales puede ser posible debido a que ya existen en el mercado sensores portátiles, inalámbricos y de bajo costo. Estos permiten registrar la actividad muscular de manera continua sin interferir con las tareas del trabajador. Estos avances son especialmente útiles en sectores como minería, manufactura, construcción y salud, donde el esfuerzo físico sostenido aumenta el riesgo de fatiga y lesiones musculoesqueléticas.

Los dispositivos actuales consideran el tamaño, la conectividad y la capacidad de procesamiento. El FreeEMG de BTS Bioengineering ofrece un sistema multicanal inalámbrico con transmisión en tiempo real y hasta 20 sondas [8]. El K-Myo de Kinvent es un sensor portátil con conectividad Bluetooth y autonomía prolongada, diseñado para análisis de fatiga y desequilibrio muscular [9]. Por su parte, el EMG One de Cometa Systems integra sensores ultraligeros con frecuencia de muestreo de hasta 2000 Hz y software para análisis en tiempo real, incluso en ambientes extremos [10], [11].

---
El *Business Understanding* nos muestra que el problema es ocupacional, social y técnico. Por lo espuesto anteriormente, la fatiga muscular debe ser detectada a tiempo para mejorar el bienestar de los trabajadores y reducir costos en las empresas.

## Data Understanding


La base de datos seleccionada parte del estudio “A Comprehensive Dataset of Surface Electromyography and Self-Perceived Fatigue Levels for Muscle Fatigue Analysis”, donde participaron 13 adultos sanos que realizaron 12 movimientos dinámicos de miembro superior (8 uniarticulares y 4 compuestos). Se emplearon sensores EMG de superficie inalámbricos para registrar la actividad de ocho músculos (cuatro por brazo: bíceps braquial y deltoides anterior, medio y posterior). Previo a los ensayos dinámicos, cada sujeto realizó tres contracciones voluntarias máximas (MVC) por músculo, empleadas como referencia para la normalización de la señal EMG. Además de los registros fisiológicos, el dataset incorpora un índice de fatiga percibida (self-perceived fatigue index), junto con datos antropométricos y de hábitos (nivel de ejercicio y consumo de cafeína).

Link de la base de datos: https://zenodo.org/records/14182446


## Características de las señales
Las señales de sEMG fueron registradas con sensores inalámbricos Delsys Trigno a 1259 Hz, con resolución de 16 bits, rango de entrada ±11 mV y un filtro pasa banda de 20–450 Hz aplicado en hardware, mientras que los índices de fatiga subjetiva se capturaron a 50 Hz de manera sincronizada. En los archivos MVC se incluye tanto la señal cruda como la procesada mediante un filtro RMS deslizante de 33 ms, mientras que en los trials se presentan las señales crudas de los músculos durante la ejecución de los movimientos. Los ensayos dinámicos se llevaron a cabo hasta que el participante alcanzara la fatiga, con descansos de 5 minutos entre series, y cada archivo incorpora marcas temporales de inicio y fin, enriqueciendo el análisis de la fatiga desde una perspectiva tanto fisiológica como subjetiva. [12]

## Graficas de la señal en el dominio del tiempo
En esta sección se presentará la señal electrica de 4 musculos del sujeto 1 (biceps braquial, deltoide anterior, medio y posterior) durante todo el tiempo de registro que fue aproximadamente de 95 segundos.

```python
#Esta parte fue solo pa probar code xd
import pandas as pd
import matplotlib.pyplot as plt

#Datos del sujeto 1
df = pd.read_csv('trial_1.csv')


def graficar_emg(x, y, label, color):
    plt.figure(figsize=(12, 6))
    plt.plot(x, y, label=label, color=color)
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Señal eléctrica (mV)')
    plt.title('Señales eléctricas del músculo')
    plt.legend()
    plt.grid(True)
    plt.show()

# Llama la función para cada señal:
graficar_emg(df['X [s]'], df['R BICEPS BRACHII: EMG 1 [V]']*1000, 'Biceps braquial', 'blue')
graficar_emg(df['X [s].1'], df['R DELTOID ANTERIOR: EMG 2 [V]']*1000, 'Deltoide anterior', 'green')
graficar_emg(df['X [s].5'], df['R DELTOID MEDIUS: EMG 6 [V]']*1000, 'Deltoide medio', 'orange')
graficar_emg(df['X [s].6'], df['R DELTOID POSTERIOR: EMG 7 [V]']*1000, 'Deltoide posterior', 'purple')
```

En las cuatro señales eléctricas musculares analizadas se observa un patrón repetitivo que corresponde a los movimientos realizados por el sujeto durante las pruebas. Este patrón es más evidente en los músculos deltoides, mientras que en el bíceps braquial resulta menos marcado. En general, no se aprecian artefactos significativos; sin embargo, en el deltoides posterior se identifican picos de voltaje que conviene considerar.

## Gráficas del espectrograma

> Se presenta los espectogramas del sujeto 1 para los 4 músculos previamente vistos. Para tener una mejor visualización se toma intervalo de los 20 primeros segundos de la señal.

```python
# se define una funcion con los parametros de los 4 espectrogramas a mostrar
def graficar_espectrograma(signal, fs, t1, t2, label, cmap='rainbow'):
    segmento = df[(df['X [s]'] >= t1) & (df['X [s]'] <= t2)]
    plt.figure(figsize=(12, 6))
    plt.specgram(segmento[signal], NFFT=512, Fs=fs, noverlap=256, cmap=cmap)
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Frecuencia (Hz)')
    plt.title(f'Espectrograma EMG {label} ({t1} a {t2} s)')
    plt.colorbar(label='Intensidad (dB)')
    plt.show()

fs = 1259  # frecuencia de muestreo que se encuentra en el articulo
t1 = 0     # tiempo inicial
t2 = 20    # tiempo final

graficar_espectrograma('R BICEPS BRACHII: EMG 1 [V]', fs, t1, t2, 'Biceps braquial')
graficar_espectrograma('R DELTOID ANTERIOR: EMG 2 [V]', fs, t1, t2, 'Deltoide anterior')
graficar_espectrograma('R DELTOID MEDIUS: EMG 6 [V]', fs, t1, t2, 'Deltoide medio')
graficar_espectrograma('R DELTOID POSTERIOR: EMG 7 [V]', fs, t1, t2, 'Deltoide posterior')
```

## Transformada Rápida de Fourier (FFT)

> Mediante la FFT se obtiene la representación espectral de la señal, mostrando cómo se distribuye la energía en diferentes frecuencias. Este análisis es útil para identificar las bandas de frecuencia predominantes en la contracción muscular e identificar ruidos.


```python
import numpy as np
from scipy.signal import spectrogram

# Columnas que corresponden a los músculos (según lo que me mostraste)
musculos = {
    "R BICEPS BRACHII": "R BICEPS BRACHII: EMG 1 [V]",
    "R DELTOID ANTERIOR": "R DELTOID ANTERIOR: EMG 2 [V]",
    "R DELTOID MEDIUS": "R DELTOID MEDIUS: EMG 6 [V]",
    "R DELTOID POSTERIOR": "R DELTOID POSTERIOR: EMG 7 [V]"
}

# Tiempo en segundos (columna X [s] de la primera serie)
tiempo = df["X [s]"]

for nombre, columna in musculos.items():
    señal = df[columna].dropna().values  # limpiar NaN
    fs = 1000  # frecuencia de muestreo en Hz (ajusta según tu data real)

    # --- FFT ---
    N = len(señal)
    fft_vals = np.fft.fft(señal)
    freqs = np.fft.fftfreq(N, 1/fs)

    plt.figure(figsize=(6,4))
    plt.plot(freqs[:N//2], np.abs(fft_vals)[:N//2], color="darkred")
    plt.title(f"FFT - {nombre}")
    plt.xlabel("Frecuencia (Hz)")
    plt.ylabel("Magnitud")
    plt.grid(True, alpha=0.3)
    plt.show()
```
