##**I. Introducción**

La electromiografía (EMG) es una técnica utilizada para registrar y analizar la actividad eléctrica generada por los músculos durante el reposo y la contracción voluntaria [1]. Para analizar esta actividad se emplean electrodos, que pueden ser de aguja o de superficie, la elección de estos depende del tipo de estudio requerido.

En este laboratorio se van a emplear electrodos de superficie, lo que se denomina electromiografía de superficie (sEMG). La sEMG es una técnica no invasiva con la que es posible detectar la suma de potenciales de acción de las fibras musculares a través de los electrodos colocados sobre la piel [2]. Es ampliamente usado para estudiar el control neuromuscular, la fatiga y los patrones de activación muscular en distintos movimientos funcionales [3]. El análisis de estas señales permite identificar fases de reposo y contracción y explorar el comportamiento en el dominio de la frecuencia o del tiempo.

###***4. Obtención de gráficas***

El BITalino cuenta con un sistema de adquisición de datos en tiempo real, lo cual es útil para registrar de manera práctica y sencilla las señales de EMG obtenidas en nuestras pruebas.  

Los datos se guardan en dos formatos principales:  
- **.h5** → archivo jerárquico (HDF5) compatible con Python y librerías como `h5py` o `pandas`.  
- **.txt** → archivo de texto delimitado, que contiene la misma información registrada pero en un formato más simple para visualización y carga rápida.  

En ambos casos, los archivos presentan 6 columnas:  

- `nSeq` → número de secuencia (contador interno).  
- `I1`, `I2` → entradas digitales.  
- `O1`, `O2` → salidas digitales.  
- `A1` → canal analógico (donde se registra la señal EMG cruda).  

Para este análisis utilizaremos los datos de la columna **A1**, ya que corresponde a las mediciones electromiográficas.  

#### Conversión de datos a Voltios  

>Los valores registrados en `A1` corresponden directamente a los códigos del ADC (10 bits), los cuales deben convertirse a voltios. La fórmula de conversión es:  
>$$
V = \frac{\text{ADC}}{2^n - 1} \times V_{ref}
$$
>Donde:  
>- $n = 10 $ (bits del ADC)  
>- $ V_{ref} = 3.3\,V$ (voltaje de referencia del BITalino).  
>
>Quedando:
>
>$$
V=\; \frac{\text{ADC}}{1023}\times 3.3
$$

#### Consideración del offset en EMG  

>La señal real de EMG tiene una amplitud muy pequeña de orden de µV a mV. El módulo de EMG del BITalino ya incluye una etapa de amplificación y filtrado, además de centrar la señal en un punto medio del ADC.  
>
>Esto significa que:  
>
>- El valor **~512** en la escala ADC corresponde aproximadamente a **1.65 V** (mitad de 3.3 V).  
>- Este punto actúa como nivel de referencia (offset) para poder representar tanto potenciales positivos como negativos de la contracción muscular.  
>- Por lo tanto, es normal que al graficar los datos la señal aparezca “desplazada hacia arriba”.  
>
>De este modo, al procesar los datos podemos restar el valor de referencia (≈512) para visualizar la señal EMG de manera centrada en torno a 0 V.

### 1. Bíceps braquial
>Se realizaron las tres condiciones de forma consecutiva (reposo, contracción dinámica ligera y contracción isométrica máxima), pero entre cada movimiento se dejó un intervalo de algunos segundos para permitir la estabilización de las señales.

```python
#@title  Gráfico completo de las 3 condiciones musculares -  Bíceps Braquial:
import h5py
import numpy as np
import matplotlib.pyplot as plt

ruta = "Biceps.h5"

with h5py.File(ruta, "r") as f:
    base = "B4:E3:F9:E4:D1:41/raw"
    # Leemos el canal A1 y aplanamos
    A1 = f[f"{base}/channel_1"][:].squeeze()

# FORZAR TIPO CON SIGNO ANTES DE RESTAR 512 (evita underflow en uint)
A1 = np.asarray(A1)
A1_signed = A1.astype(np.int32)   # ahora restar 512 es seguro

# Centrar restando 512 (porque es señal cruda)
A1_centered = A1_signed - 512

# Convertir a voltios (BITalino: Vref=3.3, 10 bits -> 1024 niveles)
Vref = 3.3
voltios_por_bit = Vref / 1024.0
A1_volt = A1_centered * voltios_por_bit

# Eje de tiempo (1000 Hz)
fs = 1000
t = np.arange(A1_volt.size) / fs

# --- GRAFICO ZOOM: cambiar inicio/fin según quieras ---
inicio = 5    # segundos
fin = 33       # segundos
idx0 = int(inicio * fs)
idx1 = int(fin * fs)
idx1 = min(idx1, A1_volt.size)

plt.figure(figsize=(12, 4))
plt.plot(t[idx0:idx1], A1_volt[idx0:idx1], color="slateblue")
plt.title("Señal completa EMG - Bíceps Braquial")
plt.xlabel("Tiempo [s]")
plt.ylabel("Voltaje [V]")
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Analisis general de la actividad del biceps braquial:
En la señal EMG registrada del bíceps braquial se aprecia claramente que entre los segundos 0 y 11 el músculo se mantiene en reposo, reflejado en un trazo de baja amplitud con valores cercanos al nivel basal. Esta fase corresponde a la actividad eléctrica mínima propia del tono muscular, sin evidencias de contracción voluntaria. En condiciones de reposo la amplitud del EMG superficial se mantiene en niveles bajos y estables [4], lo cual concuerda con lo observado en este segmento inicial.

A partir de aproximadamente el segundo 11 se evidencian incrementos notorios en la amplitud, con mayor variabilidad y picos característicos de la activación muscular voluntaria. Estos aumentos reflejan el reclutamiento progresivo de unidades motoras descrito en la literatura de De Luca [5] y confirmado en otras revisiones [6], donde se señala que la contracción voluntaria produce un patrón irregular y de amplitud elevada frente al reposo. Hacia el segundo 21 se observa un retorno transitorio a valores basales, seguido de una nueva contracción más prolongada (≈22–33 s), lo que permite distinguir fases de reposo y activación de manera clara y coherente con los patrones fisiológicos esperados en la señal EMG del bíceps.

### Analisis general del flexor radial
En la señal EMG del flexor radial del carpo se observa que, a partir de los 20 segundos, la amplitud comienza a incrementarse progresivamente, reflejando la transición del reposo hacia la contracción muscular voluntaria. A diferencia del bíceps, en este caso los periodos de reposo no están tan claramente definidos: entre los segundos 35 y 45 aproximadamente la señal mantiene cierta actividad eléctrica en lugar de volver a una línea de base estable. Esto es debido a que durante la realización del ejercicio no se tomó un descanso de más duración antes de proceder con el ejercicio de contracción isometrica.

En la segunda mitad de la señal (45–65 s) se distinguen oscilaciones de mayor amplitud y variabilidad, consistentes con contracciones sostenidas de mayor intensidad. Este patrón concuerda con lo encontrado en literatura [7], lo que señala que en músculos distales como los flexores del antebrazo la actividad EMG refleja tanto la fuerza de contracción como la participación de unidades motoras de diferentes umbrales, lo cual genera picos irregulares en la señal. La ausencia de pausas prolongadas entre esfuerzos y la variabilidad observada sugieren un reclutamiento continuo, distinto al patrón más marcado de activación y reposo visto en el bíceps.
